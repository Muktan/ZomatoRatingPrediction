```
for i,r in zomato_data.iterrows():
  zomato_data['approx_cost(for two people)'][i] = int(str(zomato_data['approx_cost(for two people)'][i]).replace(',',''))
  try:
    zomato_data['rates'][i] = float(zomato_data['rates'][i])
  except:
    zomato_data['rates'][i] = 0.0
```

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor
```
